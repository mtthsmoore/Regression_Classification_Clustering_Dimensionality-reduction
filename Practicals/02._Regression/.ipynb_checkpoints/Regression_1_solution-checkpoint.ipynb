```python
## load libraries
from sklearn import cross_validation
from sklearn.grid_search import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.svm import SVC,SVR
from sklearn import preprocessing

from sklearn.linear_model import LinearRegression,Ridge,RidgeCV
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.metrics import roc_curve,auc,precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

#Description of the data
The dataset that you will use today contains concentrations of different chemical compounds before and after a certain treatment. The first column is the initial concentration of the chemical compounds; the second column contains the gain/loss of the concentration after the treatment. Using the regression models seen in the lectures, we want to predict the gain /loss of these compounds.
The data has been spitted in training and test datasets. 

```python
# import the dataset 
data_train= pd.read_csv('regression_data_1_train.csv',sep=",")
data_test=pd.read_csv('regression_data_1_test.csv',sep=",")


## convert in numpy
train_feat= data_train['Init_Conc'].values
train_response= data_train['gain_Conc_cond1'].values

train_feat= np.reshape(train_feat,(375,1))

test_feat= data_test['Init_Conc'].values
test_response= data_test['gain_Conc_cond1'].values
test_feat= np.reshape(test_feat,(40,1))


```

We are going to evaluate different regression methods(linear and a not linear one) on our datasets.


```python
train_feat = preprocessing.scale(train_feat)

LR= LinearRegression(fit_intercept=False, normalize=False)

LR.fit(train_feat,train_response)
##
print ('Mean Square Error Training Set %.4f '
       %mean_squared_error(LR.predict(train_feat),train_response))

print ('Mean Square Error Test Set %.4f ' 
       % mean_squared_error(LR.predict(test_feat),test_response))
```

#Question 
**In regression, a good model should maximize or minimize the MSE (mean square error) **?
<p>&nbsp; </p>
<p>&nbsp; </p>
In regression a ggod model must minize the mean square error.

```python
## train a linear Regression SVM
tuned_parameters = [ {'C': np.power(2,np.arange(-8,8),dtype=float),
                      'epsilon':np.power(2,np.arange(-3,3),dtype=float) }]
# cross validate the parameter C in order to find the best model
clf = GridSearchCV(SVR(kernel='linear'),
                   tuned_parameters, cv=5, scoring='mean_squared_error')
clf.fit(train_feat, train_response)
# best C parameter
print clf.best_params_
print clf.best_estimator_
print clf.best_score_
## best MSE found
print ('Mean Square Error Training Set %.4f '
       % mean_squared_error(train_response,clf.best_estimator_.predict(train_feat)))
# we select the best model trained according to the mse
print ('Mean Square Error Test Set %.4f ' 
       % mean_squared_error(test_response,clf.best_estimator_.predict(test_feat)))
```

```python

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
f.set_size_inches(16, 7)

ax1.plot(test_feat,clf.best_estimator_.predict(test_feat),
         'b--',marker='D',label='SVM linear')
ax1.plot(test_feat,LR.predict(test_feat),'g--',marker='h',label='linear model')

ax1.scatter(test_feat,test_response,marker="*",label='True gain values')
ax1.set_title("Test Set")
ax1.set_xlabel("Initial concentration")
ax1.set_ylabel("gain_Conc_cond1")
ax1.legend(loc='upper left')


ax2.plot(train_feat,clf.best_estimator_.predict(train_feat),
         'b--',marker='D',label='SVM linear kernel')
ax2.plot(train_feat,LR.predict(train_feat),'g--',marker='h',label='linear model')
ax2.scatter(train_feat,train_response,marker="*",label='True gain values')
ax2.set_xlabel("Initial concentration")
ax2.set_ylabel("gain_Conc_cond1")
ax2.legend(loc='upper right')
ax2.set_title('Training Set')
plt.show()
```

```python
# train a rbf regression svm

tuned_parameters = [ {'C': np.power(2,np.arange(-8,8),dtype=float),
                      'gamma': np.power(2, np.arange(-8,6),dtype=float),
                      'epsilon':np.power(2,np.arange(-3,3),dtype=float) }]
# cross validate the parameter C in order to find the best model
clf_rbf = GridSearchCV(SVR(kernel='rbf'), tuned_parameters, cv=3, scoring='mean_squared_error')
clf_rbf.fit(train_feat, train_response)
# best C parameters
print clf_rbf.best_estimator_
print ('Mean Square Error Training Set %.4f '
       % mean_squared_error(train_response,clf_rbf.best_estimator_.predict(train_feat)) )
# we select the best model trained according to the mse
print ('Mean Square Error Test Set%.4f '
       % mean_squared_error(test_response,clf_rbf.best_estimator_.predict(test_feat)))
```

#Dataset Description
Load the Yeast_GeneExp.csv. The dataset contains different gene expression values measured in different conditions; some of these have been also measured on the same conditions but in different time intervals. 
We are going to train and evaluate different models in order to find which feature is better described as a combination of remaining ones. 
We fix the number of features to the first 40.

```python


data= pd.read_csv('Yeast_GeneExp.csv',sep=",")

# same preprocessing 
data= data.drop(['ORF','NAME','Label'],axis=1)

## drop the null values in the dataset
data =data.dropna()

# convert in numpy format
feat_data= data.values
## normalize the data 
feat_data = preprocessing.scale(feat_data[:,0:40])
print feat_data.shape
```

```python
## split the data in  training and test set
rs = cross_validation.ShuffleSplit(605, n_iter=1, test_size=.4,random_state=0)
for train_index, test_index in rs:
    train_set_feat = feat_data[train_index,:]
    test_set_feat = feat_data[test_index,:]
print("TRAIN:", train_set_feat.shape, "TEST:", test_set_feat.shape)
```

#Question
** What are the sizes of the training and test sets?  **

'TRAIN set' 363 sample 
'TEST set' 242 sample

```python

lm_res=[]
tuned_parameters = [ {'C': np.power(2,np.arange(-3,3),dtype=float) ,'epsilon':np.power(2,np.arange(-2,2),dtype=float)}]
ridge_res=[]
svm_res=[]
all_s= np.array(range(0,40))
for ind in range(0,40):
    print 'Feature index used as prediction' ,ind 
    print 'Train linear model'
    LR= LinearRegression(fit_intercept=False, normalize=False)
    LR.fit(train_set_feat[:,np.setdiff1d(all_s, [ind])],train_set_feat[:,ind])
    # compute the MSE on the test set
    lm_res.append(mean_squared_error(LR.predict(test_set_feat[:,np.setdiff1d(all_s, [ind])]),
                                     test_set_feat[:,[ind]] ))
    print 'Train ridge model'
    clf = RidgeCV(alphas= np.power(2,np.arange(-10,10),dtype=float) ,cv=5, scoring='mean_squared_error')
    # train a linear SVM
    # cross validate the parameter C in order to find the best model
    clf.fit(train_set_feat[:,np.setdiff1d(all_s, [ind])],train_set_feat[:,ind] )
    print 'best alpha found',clf.alpha_
    # compute the MSE on the test set
    ridge_res.append( mean_squared_error(test_set_feat[:,[ind]],
                                         clf.predict( test_set_feat[:,np.setdiff1d(all_s, [ind])]) ) )  
    #clf_lin = GridSearchCV(SVR(kernel='linear'), tuned_parameters, cv=5, scoring='mean_squared_error')
    #svm_res = mean_squared_error(test_set_feat[:,[ind]], clf_lin.best_estimator_.predict( test_set_feat[:,np.setdiff1d(all_s, [ind])]) )
    print '----'
    ##
```

```python

plt.figure(figsize=(14, 6), dpi=100)
c1=plt.plot(range(0,40),ridge_res,'g--',marker='*')
c2=plt.plot(range(0,40),lm_res,'b--',marker='h')
c3=plt.axhline(np.mean(lm_res), color='b')
c4=plt.axhline(np.mean(ridge_res), color='g')
plt.legend(('Ridge regr.','Linear regr.','Av.Linear regr.','Av. Ridge regr.'),
           ncol=3,loc='upper left')
plt.xlabel('index of the feature combination')
plt.ylabel('MSE')
plt.show()
```
